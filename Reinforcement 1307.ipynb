```python
x=np.random.normal(20, 9, 1000)
```

```python
x.std()
```

$H_0:\bar{x}=22$  
$H_1:\bar{x}\neq22$ 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(x) #distribution of data
plt.plot([22,22],[0,250]) #theoretical average that I am comparing to
plt.plot([20,20],[0,250]) #real average
plt.plot([np.quantile(x, q=0.025),np.quantile(x, q=0.025)],[0,250]) #left border
plt.plot([np.quantile(x, q=0.975),np.quantile(x, q=0.975)],[0,250]) #right border

plt.show()
```
